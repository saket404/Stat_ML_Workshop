## Plotting
Jupyter notebooks are useful for on-the-fly informal visualization. Generated plots will appear inline. 

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

FIGURE_RESOLUTION = 128
plt.rcParams['figure.dpi'] = FIGURE_RESOLUTION

import numpy as np

x = np.arange(0.0, 2.0, 0.01)
y = (x - 1)**2

plt.plot(x,y)
plt.ylabel("$y$")
plt.xlabel("$x$")
plt.title("Parabola")
plt.show()
```

## Getting Help

```python
x_1 = np.random.normal(size=1024)
x_2 = np.random.normal(loc=np.pi, scale=2, size=1024)

_ = plt.hist(x_1, bins=32, alpha=0.5)
_ = plt.hist(x_2, bins=32, alpha=0.5)
plt.ylabel('Occurrences')
plt.xlabel('$x$')
plt.show()
```
